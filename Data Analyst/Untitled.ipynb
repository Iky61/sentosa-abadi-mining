```python
import qrcode
from PIL import Image, ImageDraw, ImageFont

def create_qr_with_text(data, filename, font_size=36):
    # Buat QR Code
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    qr.add_data(data)
    qr.make(fit=True)
    qr_img = qr.make_image(fill='black', back_color='white')

    # Tentukan ukuran QR Code
    qr_width, qr_height = qr_img.size

    # Buat gambar baru dengan latar belakang putih
    img = Image.new('RGB', (qr_width, qr_height + font_size + 20), 'white')

    # Tempelkan QR Code di bagian atas gambar
    img.paste(qr_img, (0, 0))

    # Tambahkan teks di bawah QR Code dengan ukuran font yang ditentukan
    draw = ImageDraw.Draw(img)
    font = ImageFont.truetype("arial.ttf", font_size)
    text_bbox = draw.textbbox((0, 0), data, font=font)
    text_width = text_bbox[2] - text_bbox[0]
    text_height = text_bbox[3] - text_bbox[1]
    text_x = (qr_width - text_width) // 2
    text_y = qr_height + 10
    draw.text((text_x, text_y), data, font=font, fill='black')

    # Simpan gambar
    img.save(filename)

# Contoh penggunaan
data = "DT-401"
filename = "qrcode_with_text.png"
create_qr_with_text(data, filename, font_size=60)
```

```python
import qrcode
from PIL import Image, ImageDraw, ImageFont

def create_qr_with_text(data, filename, dpi=1000):
    # Buat QR Code dengan resolusi tinggi
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=20,  # Ukuran box lebih besar untuk resolusi lebih tinggi
        border=4,
    )
    qr.add_data(data)
    qr.make(fit=True)
    qr_img = qr.make_image(fill='black', back_color='white').convert('RGB')

    # Tentukan ukuran gambar
    qr_width, qr_height = qr_img.size

    # Coba untuk menggunakan arial.ttf, jika tidak ada gunakan default
    try:
        font = ImageFont.truetype("arial.ttf", 130)  # Ukuran font lebih besar
    except OSError:
        font = ImageFont.load_default()

    # Tentukan ukuran teks menggunakan getbbox
    text_bbox = font.getbbox(data)
    text_width = text_bbox[2] - text_bbox[0]
    text_height = text_bbox[3] - text_bbox[1]

    # Tentukan ukuran keseluruhan gambar
    total_height = qr_height + text_height + 60  # Tambahkan margin lebih besar
    total_width = max(qr_width, text_width)

    # Buat gambar baru dengan latar belakang putih
    img = Image.new('RGB', (total_width, total_height), 'white')

    # Gambarkan teks di atas QR Code
    draw = ImageDraw.Draw(img)
    text_x = (total_width - text_width) // 2
    text_y = 50  # Tambahkan margin atas
    draw.text((text_x, text_y), data, font=font, fill='black')

    # Tempelkan QR Code di bawah teks
    qr_x = (total_width - qr_width) // 2
    qr_y = text_y + text_height + 45  # Tambahkan margin antara teks dan QR
    img.paste(qr_img, (qr_x, qr_y))

    # Simpan gambar dengan DPI yang diinginkan
    img.save(filename, dpi=(dpi, dpi))

# Contoh penggunaan
data = "DT-401"
filename = "qrcode_with_text.png"
create_qr_with_text(data, filename, dpi=1000)
```
